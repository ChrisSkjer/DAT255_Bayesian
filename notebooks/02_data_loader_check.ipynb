# Data Loader Check

This notebook is a simple check of `bayesian_cv/data.py`.

It does three small things:
- imports the project code
- calls `load_datasets(...)`
- checks that we can read one batch from the returned datasets

The goal is not deep analysis. The goal is just to verify that the data loader works.

```python
import sys
from pathlib import Path


def find_repo_root(start: Path | None = None) -> Path:
    """Find the repository root so the notebook can import the project code."""
    current = (start or Path.cwd()).resolve()
    for candidate in [current, *current.parents]:
        if (candidate / ".git").exists() and (candidate / "bayesian_cv").exists():
            return candidate
    raise FileNotFoundError("Could not locate the repository root.")


REPO_ROOT = find_repo_root()

if str(REPO_ROOT) not in sys.path:
    sys.path.insert(0, str(REPO_ROOT))

print(f"Repository root: {REPO_ROOT}")
```

```python
from bayesian_cv.config import get_config
from bayesian_cv.data import load_datasets


# Load the datasets using the shared project config.
# This will reuse local data if it already exists, or download ImageNette if needed.
config = get_config()
train_ds, val_ds, test_ds = load_datasets(config)

print("Datasets were returned successfully.")
print(f"train_ds type: {type(train_ds)}")
print(f"val_ds type: {type(val_ds)}")
print(f"test_ds type: {type(test_ds)}")
```

```python
# Read one batch from train and validation to make sure the datasets behave as expected.
train_images, train_labels = next(iter(train_ds))
val_images, val_labels = next(iter(val_ds))

print("Train batch shapes:")
print(f"- images: {train_images.shape}")
print(f"- labels: {train_labels.shape}")

print("Validation batch shapes:")
print(f"- images: {val_images.shape}")
print(f"- labels: {val_labels.shape}")
```
